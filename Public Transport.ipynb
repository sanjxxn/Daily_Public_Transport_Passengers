```python
# full_arima_with_plots_and_csvs.py
import os
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Optional: display in notebooks
try:
    from IPython.display import display
    DISPLAY_AVAILABLE = True
except Exception:
    DISPLAY_AVAILABLE = False

# Config
FILE = r"C:\Users\HD\Desktop\Daily Public Transport\Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20250603.csv"
SERVICE_COLS = ["Local Route", "Light Rail", "Peak Service", "Rapid Route", "School"]
OUT_DIR = "simple_arima_output"
PLOT_DIR = os.path.join(OUT_DIR, "plots")
os.makedirs(OUT_DIR, exist_ok=True)
os.makedirs(PLOT_DIR, exist_ok=True)

H = 7         # forecast horizon (days)
TEST_DAYS = 7 # for plotting / quick validation

# Load & clean
df = pd.read_csv(FILE)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.set_index("Date").sort_index()

# Fill numeric gaps simply
df[SERVICE_COLS] = df[SERVICE_COLS].ffill().bfill()
if "Other" in df.columns:
    df["Other"] = df["Other"].fillna(0)

print("Data range:", df.index.min(), "to", df.index.max(), "Records:", len(df))

# Helper to show (in notebook) and save figure
def show_and_save(fig, fname):
    if DISPLAY_AVAILABLE:
        display(fig)
    fig.savefig(fname, bbox_inches="tight", dpi=150)
    plt.close(fig)

all_forecasts = {}

for col in SERVICE_COLS:
    print(f"\n=== Processing: {col} ===")
    series = df[col].astype(float).dropna()
    series.name = col

    # 1) Full historical series plot
    fig = plt.figure(figsize=(12,3))
    plt.plot(series, marker='.', linewidth=0.6)
    plt.title(f"{col}  Full Time Series")
    plt.grid(True)
    fname = os.path.join(PLOT_DIR, f"{col.replace(' ','_')}_full_series.png")
    show_and_save(fig, fname)
    print("Saved:", fname)

    # If too short -> naive forecast
    if len(series) < 20:
        print("Series too short for ARIMA; using naive forecast.")
        last = series.iloc[-1]
        future = pd.Series([last]*H, index=pd.date_range(start=series.index[-1] + pd.Timedelta(days=1), periods=H, freq='D'))
        all_forecasts[col] = future.round(1)

        # Plot naive forecast
        fig = plt.figure(figsize=(10,3))
        plt.plot(series[-60:], label='Last 60 days', color='gray')
        plt.plot(future.index, future.values, marker='o', label='7-day forecast', color='green')
        plt.title(f"{col}  7-Day Forecast (Naive)")
        plt.legend(); plt.grid(True)
        fname = os.path.join(PLOT_DIR, f"{col.replace(' ','_')}_7day_forecast_naive.png")
        show_and_save(fig, fname)
        print("Saved:", fname)

        # Save CSV
        future.to_csv(os.path.join(OUT_DIR, f"{col.replace(' ','_')}_7day_arima.csv"), header=[col])
        print("Saved CSV:", os.path.join(OUT_DIR, f"{col.replace(' ','_')}_7day_arima.csv"))
        continue  # next column

    # 2) Train/test split
    train = series.iloc[:-TEST_DAYS]
    test = series.iloc[-TEST_DAYS:]

    # 3) Fit ARIMA with fallback
    try:
        model = ARIMA(train, order=(2,1,2))
        fit = model.fit()
    except Exception as e:
        print("ARIMA(2,1,2) failed, falling back to ARIMA(1,1,1). Error:", e)
        model = ARIMA(train, order=(1,1,1))
        fit = model.fit()

    # 4) Predict test period
    pred_test = fit.forecast(steps=len(test))
    pred_test.index = test.index

    # 5) Plot Train/Test/Predicted
    fig = plt.figure(figsize=(12,3))
    plt.plot(train.index, train, label='Train', color='gray')
    plt.plot(test.index, test, label='Actual (Test)', color='orange')
    plt.plot(pred_test.index, pred_test, label='Predicted', color='red')
    plt.title(f"{col}  Train / Test / Predicted")
    plt.legend(); plt.grid(True)
    fname = os.path.join(PLOT_DIR, f"{col.replace(' ','_')}_train_test_pred.png")
    show_and_save(fig, fname)
    print("Saved:", fname)

    # 6) Forecast next H days
    future = fit.forecast(steps=H)
    future.index = pd.date_range(start=series.index[-1] + pd.Timedelta(days=1), periods=H, freq='D')
    all_forecasts[col] = future.round(1)

    # 7) Plot future forecast with recent context
    fig = plt.figure(figsize=(10,3))
    plt.plot(series[-60:], label='Last 60 days', color='gray')
    plt.plot(future.index, future.values, marker='o', label='7-day forecast', color='green')
    plt.title(f"{col}  7-Day Forecast")
    plt.legend(); plt.grid(True)
    fname = os.path.join(PLOT_DIR, f"{col.replace(' ','_')}_7day_forecast.png")
    show_and_save(fig, fname)
    print("Saved:", fname)

    # 8) Save per-column CSVs
    future.to_csv(os.path.join(OUT_DIR, f"{col.replace(' ','_')}_7day_arima.csv"), header=[col])
    pred_test.to_csv(os.path.join(OUT_DIR, f"{col.replace(' ','_')}_pred_test.csv"), header=[col + "_pred"])
    print("Saved CSVs:", os.path.join(OUT_DIR, f"{col.replace(' ','_')}_7day_arima.csv"),
          ",", os.path.join(OUT_DIR, f"{col.replace(' ','_')}_pred_test.csv"))

    # 9) Print quick metrics for the test window (optional)
    try:
        from sklearn.metrics import mean_absolute_error, mean_squared_error
        mae = mean_absolute_error(test, pred_test)
        rmse = mean_squared_error(test, pred_test, squared=False)
        print(f"MAE (last {len(test)} days): {mae:.2f}, RMSE: {rmse:.2f}")
    except Exception:
        pass

# After loop: save combined forecast table
combined = pd.DataFrame(all_forecasts)
combined.index.name = "date"
combined.to_csv(os.path.join(OUT_DIR, "all_services_7day_arima.csv"))
print("\nAll forecasts saved in:", OUT_DIR)
print("Saved plot images to:", PLOT_DIR)
print("\nCombined 7-day forecast (preview):")
print(combined.head(H))
```
